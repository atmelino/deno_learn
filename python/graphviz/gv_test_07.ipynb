```python
from graphviz import Digraph
dot = Digraph(format="svg", node_attr={"shape": "record"}, graph_attr={"rankdir": "LR"})  # LR = left to right

with dot.subgraph(name="cluster_i") as c:
    c.attr(style="filled", color="lightgrey")
    c.node_attr.update(style="filled", color="white")
    c.node("i1")
    c.attr(label="layer i")


with dot.subgraph(name="cluster_2") as c:
    c.attr(color="blue")
    c.node_attr["style"] = "filled"
    c.node(
        "n3",
        shape="record",
    )

    c.node(
        name="fr",
        label="{ %s |%s |%s |%s | data %.4f | grad %.4f }"
        % ("a","b", "c", "d", 3.4 , 0.73),
        shape="record",
    )

    c.node(
        name="fr2",
        label="{ %s | data %.4f  }"
        % ( "d", 0.73),
        shape="record",
    )

    c.node(
        "n4",
        shape="record",
    )
    c.attr(label="layer #2")

with dot.subgraph(name="cluster_o") as c:
    c.attr(style="filled", color="lightgrey")
    c.node_attr.update(style="filled", color="white")
    c.node("o1")
    c.attr(label="layer o")



with dot.subgraph(name="cluster_6") as c:
    c.attr(style="filled", color="lightgrey")
    c.node_attr.update(style="filled", color="white")
    c.attr(label="layer i")
    c.node("N1", r"  N1 |{w1 | 0.64} |{w2 | 0.53}|{b | 0.53}")
    c.node("N2", r"  N2 |{w1 | 0.64} |{w2 | 0.53}|{b | 0.53}")



dot.edge("i1", "N1")
dot.edge("i1", "N2")
dot.edge("n1", "n3")
dot.edge("n2", "n3")
dot.edge("n1", "n4")
dot.edge("n2", "n4")
dot.edge("n3", "o1")
dot.edge("n4", "o1")

dot
```
