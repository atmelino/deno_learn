```python
import graphviz

s = graphviz.Digraph(
    "structs", filename="structs_revisited.gv", node_attr={"shape": "record"}, graph_attr={"rankdir": "LR"}
)

with s.subgraph(name="cluster_i") as c:
    c.attr(style="filled", color="lightgrey")
    c.node_attr.update(style="filled", color="white")
    c.attr(label="layer i")
    c.node("N1", r"  N1 |{w1 | 0.64} |{w2 | 0.53}|{b | 0.53}")
    c.node("N2", r"  N2 |{w1 | 0.64} |{w2 | 0.53}|{b | 0.53}")


s
```
