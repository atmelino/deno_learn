```python
from graphviz import Digraph

dot = Digraph(
    format="svg", node_attr={"shape": "record"}, graph_attr={"rankdir": "LR"}
)  # LR = left to right

with dot.subgraph(name="cluster_i") as c:
    c.attr(label="layer i")
    c.attr(style="filled", color="lightgrey")
    c.node_attr.update(style="filled", color="white")
    c.node("i1")

with dot.subgraph(name="cluster_6") as c:
    c.attr(label="layer 1")
    c.attr(style="filled", color="lightgrey")
    c.node_attr.update(style="filled", color="white")
    c.node(
        "N1",
        r"  %s |{%s |d %.4f g %.4f} |{%s |d %.4f g %.4f}|{b |d %.4f g %.4f}"
        % ("N1", "w1", 0.64, -0.2, "w2", 0.53, -0.31, 0.18, 0.22),
    )
    c.node("N2", r"  N2 |{w1 | 0.64} |{w2 | 0.53}|{b | 0.53}")

with dot.subgraph(name="cluster_o") as c:
    c.attr(label="layer o")
    c.attr(style="filled", color="lightgrey")
    c.node_attr.update(style="filled", color="white")
    c.node("o1")


dot.edge("i1", "N1")
dot.edge("i1", "N2")
dot.edge("N1", "o1")
dot.edge("N2", "o1")

dot
```
