```python
"""https://www.graphviz.org/pdf/dotguide.pdf, Figure 12"""

import graphviz

s = graphviz.Digraph(
    "structs", filename="structs_revisited.gv", node_attr={"shape": "record"}
)


with s.subgraph(name="cluster_i") as c:
    c.attr(style="filled", color="lightgrey")
    c.node_attr.update(style="filled", color="white")
    c.node("i1")
    c.attr(label="layer i")
    c.node("struct3", r"hello\nworld |{ b |{c|<here> d|e}| f}| g | h")

s
```
