```python
import math
import numpy as np
import matplotlib.pyplot as plt
from graphviz import Digraph

%matplotlib inline
```

```python
# dot = Digraph(
#     format="svg", graph_attr={"rankdir": "LR"}, node_attr={"shape": "plaintext"}
# )  # LR = left to right

dot = Digraph(format="svg", graph_attr={"rankdir": "LR"})  # LR = left to right

# NOTE: the subgraph name needs to begin with 'cluster' (all lowercase)
#       so that Graphviz recognizes it as a special cluster subgraph

with dot.subgraph(name="cluster_i") as c:
    c.attr(style="filled", color="lightgrey")
    c.node_attr.update(style="filled", color="white")
    c.node("i1")
    c.attr(label="layer i")

with dot.subgraph(name="cluster_1") as c:
    c.attr(style="filled", color="lightgrey")
    c.node_attr.update(style="filled", color="white")
    c.node(
        "n1",
        shape="record",
    )
    c.node("struct1", "<f0> left|<f1> middle|<f2> right")
    c.node("struct2", "<f0> one|<f1> two")
    c.node("struct3", r"hello\nworld |{ b |{c|<here> d|e}| f}| g | h")

    c.node("na", "w1 | 0.34\nw2 | 0.5")

    c.node("na", "w1\nw2 | 0.34\n 0.5", shape="record")

    c.node(
        "n2",
        shape="record",
    )
    c.attr(label="layer #1")

with dot.subgraph(name="cluster_2") as c:
    c.attr(color="blue")
    c.node_attr["style"] = "filled"
    c.node(
        "n3",
        shape="record",
    )

    c.node(
        name="fr",
        label="{ %s |%s |%s |%s | data %.4f | grad %.4f }"
        % ("a","b", "c", "d", 3.4 , 0.73),
        shape="record",
    )


    c.node(
        "n4",
        shape="record",
    )
    c.attr(label="layer #2")

with dot.subgraph(name="cluster_o") as c:
    c.attr(style="filled", color="lightgrey")
    c.node_attr.update(style="filled", color="white")
    c.node("o1")
    c.attr(label="layer o")

dot.edge("i1", "n1")
dot.edge("i1", "n2")
dot.edge("n1", "n3")
dot.edge("n2", "n3")
dot.edge("n1", "n4")
dot.edge("n2", "n4")
dot.edge("n3", "o1")
dot.edge("n4", "o1")

dot
```
